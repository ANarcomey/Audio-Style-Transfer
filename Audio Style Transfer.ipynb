```python
contentSpectro, content_sr = read_audio_spectrum(content_file, N_FFT = 2048)
styleSpectro, style_sr = read_audio_spectrum(style_file, N_FFT = 2048)

N_binsC, N_timestepsC = contentSpectro.shape
N_binsS, N_timestepsS = styleSpectro.shape

print("content spectrogram shape = ", contentSpectro.shape)
print("style spectrogram shape = ", styleSpectro.shape)
```

```python
#ensure that content and style have the same shape, though shouldn't be an issue is clips
#are cut to same length
styleSpectro = styleSpectro[:N_binsC,:N_timestepsC]
```

```python
## Create tensors for content and style by reshaping spectrograms

reshapedContent = np.reshape(contentSpectro.T, (1,1,N_timestepsC, N_binsC))
reshapedStyle = np.reshape(styleSpectro.T, (1,1,N_timestepsS, N_binsS))

content_tf = tf.constant(reshapedContent, name = 'content_tf', dtype = tf.float32)
style_tf = tf.constant(reshapedStyle, name = 'style_tf', dtype = tf.float32)
```
