```python
import librosa
import librosa.display
import numpy as np

import matplotlib.pyplot as plt
```

```python
N_FFT = 2048
wav, sampleRate = librosa.load("savedOutputs/NewReconstruction/4096EqualContentWeightEGuitar/out.wav")
spectrogram = librosa.stft(wav, N_FFT)
```

```python
librosa.display.specshow(librosa.amplitude_to_db(spectrogram, ref=np.max), y_axis='log', x_axis='time')
plt.title('Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
```
